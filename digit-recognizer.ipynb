```python
test_data = pd.read_csv('/kaggle/input/digit-recognizer/test.csv')
train_data = pd.read_csv('/kaggle/input/digit-recognizer/train.csv')
```

```python
train_data
```

```python
y = train_data.loc[:, 'label']
X = train_data.iloc[:, 1:]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Neural Network

```python
from sklearn.neural_network import MLPClassifier

nn_clf = MLPClassifier(hidden_layer_sizes=(100,50), activation='logistic', 
                       solver = 'sgd', learning_rate='adaptive', random_state=42)
nn_clf.fit(X_train, y_train)
nn_clf.predict(X_test)
nn_clf.score(X_test, y_test)
```

```python
test_data
```

```python
output_predictions = nn_clf.predict(test_data)
output_predictions
```

```python
output = pd.DataFrame({
    'ImageId': np.arange(1, len(output_predictions) + 1),
    'Label' : output_predictions})
output.to_csv('NN Model', index=False)
```
